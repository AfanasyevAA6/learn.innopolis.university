```python
# загрузим датасет
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Titanic-Dataset.csv')
df
```

```python
# анализ
analys = pd.DataFrame(index = df.keys())

analys['unique'] = [len(df[i].unique()) for i in df.keys()]
analys['min'] = df.describe().loc['min']
analys['max'] = df.describe().loc['max']
analys
```

```python
analys['unique'].tolist()
```

```python
import matplotlib.pyplot as plt
```

```python
# Создание гистограммы
plt.bar(df.keys(), analys['unique'].tolist(), label = 'Unique number')

# Настройка осей и заголовка
plt.title('Количество уникальных значений')
plt.xlabel('Параметры')
plt.ylabel('Кол-во уникальных значений')
plt.xticks(fontsize=8)
plt.xticks(rotation=45)

# добавление легенды
plt.legend()

# Отображение графика
plt.show()
```
