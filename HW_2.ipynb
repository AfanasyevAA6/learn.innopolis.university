Проверяем гипотезу о том, что метод случайного леса прогнозирует лучше чем метод линейной регрессии.

```python
import scipy.stats as stats
import pandas as pd
import numpy as np
```

Получаем подтверждение высокой валидности примененного нами препроцессинга

```python
df = pd.read_excel('ML_result.xlsx')
df
```

В таблице соотносится прогнозируемый параметр и метрика MAPE прогнозирования модели. Данные имеют в некоторых строках нечисленные значение. Следовательно, необходимо от них избавится.

```python
#delete (NULL.inf-infNaN)
df = df[df.replace([np.inf, -np.inf], np.nan).notnull().all(axis=1)]  # .astype(np.float64) 
df
```

```python
A = df['MAPE (Linear)']
B = df['MAPE (Decision Tree)']
print(A, B)
```

```python
# Perform paired samples t-test
p = stats.ttest_rel(A, B)
statics, p_value = stats.ttest_rel(A, B)
print(p)
# Print the results
print("p-value:", p_value, '\nstatistic=', statics)
```

Значениер   —   это вероятность получения нашего вычисленного значения критерия или его еще большего значения,   если нулевая гипотеза верна. Иными словами,  p  -  это вероятность отвергнуть нулевую гипотезу при условии, что она верна.
